```python
from svgpathtools import parse_path, svg2paths

# SVG 파일을 읽어옵니다.
paths, _ = svg2paths('서울_시군구.svg')

# 각 경로에서 선(segment)을 추출하고 점들의 좌표를 얻습니다.
for path in paths:
    for segment in path:
        # 선(segment)을 따라 점들의 좌표를 얻습니다.
        points = parse_path(segment)
        for point in points:
            # 점의 좌표를 출력하거나 원하는 작업을 수행합니다.
            print(point)
```

```python
from svgpathtools import parse_path, svg2paths

# SVG  .
paths, _ = svg2paths('_.svg')

#   (segment)    .
for path in paths:
    for segment in path:
        # (segment)    .
        points = parse_path(segment)
        for point in points:
            #      .
            print(point)
```

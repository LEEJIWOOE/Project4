## 테스트 1
> 이미지를 잘 가져오나 슬라이드의 이미지 수를 수동으로 지정해놓음.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import requests
import os
import time

# 설정 옵션: 크롬 브라우저를 분리하여 실행
options = Options()
options.add_experimental_option("detach", True)
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 목표 URL
url = "https://www.recycling-info.or.kr/act4r/info/tip/list.do#plastic/30"

# 웹페이지 열기
driver.get(url)

# 페이지 로딩 대기
time.sleep(3)  # 페이지가 로드될 때까지 3초 대기

# 슬라이드 이미지를 추출하는 함수
def get_slide_images(driver):
    images = []
    slide_index = 0
    
    while True:
        try:
            # 슬라이드 이미지가 로드될 때까지 대기
            WebDriverWait(driver, 1).until(
                lambda driver: driver.find_element(By.CSS_SELECTOR, "div.view-box img")
            )
            
            # 슬라이드 이미지 요소를 찾기
            img_elements = driver.find_elements(By.CSS_SELECTOR, "div.view-box img")
            for img_element in img_elements:
                img_url = img_element.get_attribute("src")
                if img_url not in images:
                    images.append(img_url)
            
            # 'Next' 버튼 클릭
            next_button = driver.find_element(By.CSS_SELECTOR, "button.slick-next.slick-arrow")
            next_button.click()
            
            time.sleep(1)  # 슬라이드가 넘어갈 시간을 줍니다.
            slide_index += 1

            # 최대 슬라이드 수를 제한하거나 종료 조건을 설정할 수 있음
            # if slide_index > 10:  # 임시로 10개 슬라이드로 제한
                # break
        except Exception as e:
            print(f"슬라이드 이미지를 찾을 수 없습니다: {e}")
            break
    
    return images

# 슬라이드에서 모든 이미지 URL 가져오기
img_urls = get_slide_images(driver)

# 디렉토리 생성 (존재하지 않으면)
os.makedirs('img', exist_ok=True)

# 이미지 다운로드 및 저장
for idx, img_url in enumerate(img_urls):
    save_file_name = f'img/view_box_image_{idx}.jpg'
    img_data = requests.get(img_url).content
    with open(save_file_name, "wb") as img_file:
        img_file.write(img_data)
    print(f"Image saved as {save_file_name}")

driver.quit()
```

## 테스트 2
> 이미지의 URL이 같으면 종료하는 방식으로 무한루프를 해결하고 유동적인 이미지개수를 전부가져오도록 수정함
> 그러나 이미지의 URL이 전부 같은걸 확인하고 첫번째 이미지를 가져오고 두번째 이미지에서 종료되는 오류 발생!

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import requests
import os
import time

# 설정 옵션: 크롬 브라우저를 분리하여 실행
options = Options()
options.add_experimental_option("detach", True)
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 목표 URL
url = "https://www.recycling-info.or.kr/act4r/info/tip/list.do#plastic/31"

# 웹페이지 열기
driver.get(url)

# 페이지 로딩 대기
time.sleep(3)  # 페이지가 로드될 때까지 3초 대기

# 슬라이드 이미지를 추출하는 함수
def get_slide_images(driver):
    images = []
    
    while True:
        try:
            # 현재 슬라이드 이미지가 로드될 때까지 대기
            WebDriverWait(driver, 10).until(
                lambda driver: driver.find_element(By.CSS_SELECTOR, "div.view-box img")
            )
            
            # 현재 슬라이드 이미지 요소를 찾기
            img_element = driver.find_element(By.CSS_SELECTOR, "div.view-box img")
            img_url = img_element.get_attribute("src")
            if img_url not in images:
                images.append(img_url)
            
            # 'Next' 버튼 클릭
            next_button = driver.find_element(By.CSS_SELECTOR, "button.slick-next.slick-arrow")
            next_button.click()
            
            time.sleep(1)  # 슬라이드가 넘어갈 시간을 줍니다.

            # 다음 슬라이드로 넘어가면 첫 번째 이미지와 현재 이미지가 같은지 확인하여 슬라이드가 끝났는지 판단
            first_img_element = driver.find_element(By.CSS_SELECTOR, "div.view-box img")
            first_img_url = first_img_element.get_attribute("src")
            if first_img_url == img_url:
                break  # 슬라이드가 끝났으면 루프 종료
        except Exception as e:
            print(f"슬라이드 이미지를 찾을 수 없습니다: {e}")
            break
    
    return images

# 슬라이드에서 모든 이미지 URL 가져오기
img_urls = get_slide_images(driver)

# 디렉토리 생성 (존재하지 않으면)
os.makedirs('img', exist_ok=True)

# 이미지 다운로드 및 저장
for idx, img_url in enumerate(img_urls):
    save_file_name = f'img/view_box_image_{idx}.jpg'
    img_data = requests.get(img_url).content
    with open(save_file_name, "wb") as img_file:
        img_file.write(img_data)
    print(f"Image saved as {save_file_name}")

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os
import requests

# 이미지를 저장하는 함수
def save_image(image_url, filename):
    try:
        # 이미지 다운로드
        img_data = requests.get(image_url).content
        # 이미지 저장
        with open(filename, "wb") as img_file:
            img_file.write(img_data)
        print(f"{filename} 저장 완료")
    except Exception as e:
        print(f"이미지 저장 중 오류 발생: {e}")

# 이미지를 가져오는 함수
def get_slide_images(driver):
    images = []
    seen_images = set()  # 이미지 중복 확인을 위한 집합
    
    try:
        # 슬라이드 이미지가 로드될 때까지 대기
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "div.view-box img"))
        )
        
        # 모든 슬라이드 이미지 요소를 가져오기
        img_elements = driver.find_elements(By.CSS_SELECTOR, "div.view-box img")
        
        # 각 이미지의 src 속성을 가져와서 리스트에 추가
        for img_element in img_elements:
            img_url = img_element.get_attribute("src")
            
            # 중복된 이미지인지 확인하여 중복 제거
            if img_url not in seen_images:
                images.append(img_url)
                seen_images.add(img_url)
            
    except Exception as e:
        print(f"슬라이드 이미지를 찾을 수 없습니다: {e}")
    
    return images

# 반복문을 통해 URL 생성
for i in range(1, 5):
    # URL의 마지막 숫자를 변경하여 새 URL 생성
    url = f"https://www.recycling-info.or.kr/act4r/info/tip/list.do#plastic/{i}"
    
    # 웹 드라이버 시작
    driver = webdriver.Chrome()
    driver.get(url)
    
    # 이미지 가져오기
    images = get_slide_images(driver)
    
    # 각 이미지를 img 폴더에 저장
    for idx, img_url in enumerate(images):
        filename = f"img/image_{i}_{idx}.jpg"  # 이미지 파일명 생성
        save_image(img_url, filename)
    
    # 웹 드라이버 종료
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os
import requests

# 이미지를 저장하는 함수
def save_image(image_url, filename):
    try:
        # 이미지 다운로드
        img_data = requests.get(image_url).content
        # 이미지 저장
        with open(filename, "wb") as img_file:
            img_file.write(img_data)
        print(f"{filename} 저장 완료")
    except Exception as e:
        print(f"이미지 저장 중 오류 발생: {e}")

# 이미지를 가져오는 함수
def get_slide_images(driver):
    images = []
    seen_images = set()  # 이미지 중복 확인을 위한 집합
    
    try:
        # 슬라이드 이미지가 로드될 때까지 대기
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "div.view-box img"))
        )
        
        # 모든 슬라이드 이미지 요소를 가져오기
        img_elements = driver.find_elements(By.CSS_SELECTOR, "div.view-box img")
        
        # 각 이미지의 src 속성을 가져와서 리스트에 추가
        for img_element in img_elements:
            img_url = img_element.get_attribute("src")
            
            # 중복된 이미지인지 확인하여 중복 제거
            if img_url not in seen_images:
                images.append(img_url)
                seen_images.add(img_url)
            
    except Exception as e:
        print(f"슬라이드 이미지를 찾을 수 없습니다: {e}")
    
    return images

# 숫자 리스트
plastic = [30,31,40,39,33,32,38,34,35,37,29]
disposable = [36,22,13,1]
waste_elec = [18,28,27,26]
food_waste = [8,23,11,12]
waste = [21,2,24]

# 각 이미지에 해당하는 숫자와 폴더 이름을 매핑한 딕셔너리 생성
image_folders = {
    0: "분리배출",
    1: "일회용품",
    2: "폐가전",
    3: "음식물폐기물",
    4: "기타폐기물"
}

# 각 리스트에 대해 중첩된 for 루프 사용하여 각각의 경우에 대한 작업 수행
for a in plastic:
    for b in disposable:
        for c in waste_elec:
            for d in food_waste:
                for e in waste:
                    # URL 생성
                    url = f"https://www.recycling-info.or.kr/act4r/info/tip/list.do#plastic/{a}"
                    
                    # 웹 드라이버 시작
                    driver = webdriver.Chrome()
                    driver.get(url)
                    
                    # 이미지 가져오기
                    images = get_slide_images(driver)
                    
                    # 각 이미지를 해당 폴더에 저장
                    for idx, img_url in enumerate(images):
                        # 폴더별로 파일 경로 설정
                        folder_name = image_folders.get(idx, "기타폐기물")  # 기본값은 "기타폐기물"
                        folder_path = os.path.join("img", folder_name)
                        os.makedirs(folder_path, exist_ok=True)  # 폴더가 없으면 생성
                        filename = f"image_{a}{idx}.jpg"  # 이미지 파일명 생성
                        save_image(img_url, os.path.join(folder_path, filename))  # 폴더 경로를 포함한 파일 경로로 수정
                    
# 웹 드라이버 종료
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os
import requests
import re

# 이미지를 저장하는 함수
def save_image(image_url, filename):
    try:
        img_data = requests.get(image_url).content
        with open(filename, "wb") as img_file:
            img_file.write(img_data)
        print(f"{filename} 저장 완료")
    except Exception as e:
        print(f"이미지 저장 중 오류 발생: {e}")

# 이미지를 가져오는 함수
def get_slide_images(driver):
    images = []
    seen_images = set()

    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "div.view-box img"))
        )

        img_elements = driver.find_elements(By.CSS_SELECTOR, "div.view-box img")

        for img_element in img_elements:
            img_url = img_element.get_attribute("src")
            if img_url not in seen_images:
                images.append(img_url)
                seen_images.add(img_url)

    except Exception as e:
        print(f"슬라이드 이미지를 찾을 수 없습니다: {e}")

    return images

# URL에서 숫자 부분 추출 함수
def extract_numbers(url):
    pattern = r"#(\w+)/(\d+)"
    match = re.search(pattern, url)
    if match:
        category = match.group(1)
        number = int(match.group(2))
        return category, number
    else:
        return None, None

# 섹션별 URL 및 숫자 리스트
sections = {
    "plastic": [30, 31, 40, 39, 33, 32, 38, 34, 35, 37, 29],
    "disposable": [36, 22, 13, 1],
    "waste-elec": [18, 28, 27, 26],
    "food-waste": [8, 23, 11, 12],
    "waste": [21, 2, 24]
}

# 폴더명 매핑
folder_names = {
    "plastic": "분리배출",
    "disposable": "일회용품",
    "waste-elec": "폐가전",
    "food-waste": "음식물폐기물",
    "waste": "기타폐기물"
}

# 반복문을 통해 각 섹션의 이미지 저장 및 숫자별 폴더 생성
for section, numbers in sections.items():
    folder_name = folder_names[section]
    section_folder = f"img/{folder_name}"
    os.makedirs(section_folder, exist_ok=True)

    for number in numbers:
        url = f"https://www.recycling-info.or.kr/act4r/info/tip/list.do#{section}/{number}"
        driver = webdriver.Chrome()
        driver.get(url)

        images = get_slide_images(driver)

        number_folder = f"{section_folder}/{number}"
        os.makedirs(number_folder, exist_ok=True)

        for idx, img_url in enumerate(images):
            filename = f"{number_folder}/image_{idx}.jpg"
            save_image(img_url, filename)

        driver.quit()

print("작업이 완료되었습니다.")
```
